```python
import requests

def get_top_starred_repos():
    # Set API endpoint and parameters
    url = "https://api.github.com/search/repositories"
    params = {
        "q": "stars:>1000",  # Search for repositories with more than 1000 stars
        "sort": "stars",  # Sort by star count in descending order
        "order": "desc",
        "per_page": 100  # Fetch 100 repositories per page
    }
    
    top_repos = []
    page = 1
    
    while len(top_repos) < 1000:
        params['page'] = page
        response = requests.get(url, params=params)
        
        if response.status_code == 200:
            data = response.json()
            if data['items']:
                top_repos.extend(data['items'])
                if len(data['items']) < params['per_page']:
                    break  # No more pages
            else:
                print("No more repositories found.")
                break
        else:
            print(f"Failed to retrieve repositories. Status code: {response.status_code}")
            break
        
        page += 1
    
    # Trim the list to the top 1000 repositories
    top_repos = top_repos[:1000]
    
    return top_repos

# Example usage
top_starred_repos = get_top_starred_repos()

for i, repo in enumerate(top_starred_repos):
    print(f"Rank {i+1}: {repo['name']} - Stars: {repo['stargazers_count']}")
```
