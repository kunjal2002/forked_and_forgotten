```python
import pandas as pd
import matplotlib.pyplot as plt

# Example mean lifespan data for three repositories
repo_names = [
    'freeCodeCamp/freeCodeCamp',
    'EbookFoundation/free-programming-books',
    'sindresorhus/awesome'
]

# Replace these values with the actual mean lifespan for each repository
mean_lifespan_months = [mean_lifespan_months_1, mean_lifespan_months_2, mean_lifespan_months_3]
mean_lifespan_years = [mean_lifespan_years_1, mean_lifespan_years_2, mean_lifespan_years_3]

# Create DataFrame for plotting
data_months = pd.DataFrame({
    'Repository': repo_names,
    'Mean Lifespan (months)': mean_lifespan_months
})

data_years = pd.DataFrame({
    'Repository': repo_names,
    'Mean Lifespan (years)': mean_lifespan_years
})

# Function to add labels on bars
def add_labels_bars(rects, ax):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom',
                    fontweight='bold', fontsize=10)

# Plotting Mean Lifespan in Months
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(data_months['Repository'], data_months['Mean Lifespan (months)'], color='navy', alpha=0.8)
add_labels_bars(bars, ax)
ax.set_xlabel('Repository', fontsize=12, fontweight='bold', fontname='Arial')
ax.set_ylabel('Average Lifespan (months)', fontsize=12, fontweight='bold', fontname='Arial')
ax.set_title('Average Lifespan in Months for Repositories', fontsize=14, fontweight='bold', fontname='Arial')
ax.set_xticklabels(data_months['Repository'], rotation=0, fontweight='bold', fontsize=10, fontname='Arial')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.margins(y=0.1)  # Add margins to avoid cutting off the numbers
plt.show()

# Plotting Mean Lifespan in Years
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(data_years['Repository'], data_years['Mean Lifespan (years)'], color='navy', alpha=0.8)
add_labels_bars(bars, ax)
ax.set_xlabel('Repository', fontsize=12, fontweight='bold', fontname='Arial')
ax.set_ylabel('Mean Lifespan (years)', fontsize=12, fontweight='bold', fontname='Arial')
ax.set_title('Mean Lifespan in Years for Repositories', fontsize=14, fontweight='bold', fontname='Arial')
ax.set_xticklabels(data_years['Repository'], rotation=0, fontweight='bold', fontsize=10, fontname='Arial')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.margins(y=0.1)  # Add margins to avoid cutting off the numbers
plt.show()
```
