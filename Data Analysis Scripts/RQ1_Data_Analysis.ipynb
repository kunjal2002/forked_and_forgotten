```python
import pandas as pd
```

```python
data_file=pd.read_excel("github_fetched_contributor_data.xlsx")
```

```python
df=pd.DataFrame(data_file)
```

```python
df.head(10)
```

```python
#Dropping unneccesary columns as per RQ1
df.drop(['First Contribution', 'Last Contribution', 'Frequency', 'Status', 'First Activity', 'Last Activity'], axis=1, inplace=True)
```

```python
df.head()
```

```python
mean_lifespans = df.groupby('Repository Name')['Lifespan (days)'].mean()
```

```python
print(mean_lifespans)
mean_lifespans.info()
```

```python
import matplotlib.pyplot as plt

# Assuming mean_lifespans is a Pandas Series
plt.figure(figsize=(15, 6))  # Adjust size for better readability
mean_lifespans.plot(kind='bar', color='#03055B')  # Use a dark blue color
plt.xlabel('Repository Name', fontsize=14, fontweight='bold')  # Make x-axis label bold
plt.ylabel('Mean Lifespan (In days)', fontsize=14, fontweight='bold')  # Make y-axis label bold
plt.title('Mean Lifespan by Repository', fontsize=16, fontweight='bold')  # Make title bold
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels and align them to the right
plt.tight_layout()  # Ensure labels fit within the figure
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid for better readability
plt.show()
```

```python
#Check Correlation
repo_data=pd.read_csv("repositories_info.csv")
```

```python
repo_data_df=pd.DataFrame(repo_data)
```

```python
repo_data_df.head()
```

```python
repo_data_df['Mean Lifespan'] = repo_data_df['Repository Name'].map(mean_lifespans)
```

```python
repo_data_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlation matrix
correlation_matrix = repo_data_df[['Stars', 'Forks', 'Total Issues', 'Total Pull Requests', 'Total Files', 'Lines of Code', 'Contributors', 'Open Issues', 'Closed Issues', 'Issue Resolution Ratio (%)', 'Mean Lifespan']].corr()

# Visualize correlation matrix
plt.figure(figsize=(12,10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)
plt.show()

# Example scatter plot for one feature
plt.scatter(repo_data_df['Stars'], repo_data_df['Mean Lifespan'])
plt.xlabel('Stars')
plt.ylabel('Mean Lifespan')
plt.title('Correlation between Stars and Mean Lifespan')
plt.show()
```

```python
correlation_matrix
```

```python
from scipy.stats import pearsonr

# Calculate correlation coefficient and p-value for Stars vs. Mean Lifespan
r_stars, p_stars = pearsonr(repo_data_df['Stars'], repo_data_df['Mean Lifespan'])
print(f"Correlation between Stars and Mean Lifespan: r = {r_stars}, p = {p_stars}")

# Calculate correlation coefficient and p-value for Contributors vs. Mean Lifespan
r_contributors, p_contributors = pearsonr(repo_data_df['Contributors'], repo_data_df['Mean Lifespan'])
print(f"Correlation between Contributors and Mean Lifespan: r = {r_contributors}, p = {p_contributors}")
```

```python
from scipy.stats import mannwhitneyu

# Define the groups
small_contributors = repo_data_df[repo_data_df['Contributors'] < 50]['Mean Lifespan']
large_contributors = repo_data_df[repo_data_df['Contributors'] >= 50]['Mean Lifespan']

# Perform the Mann-Whitney U test
u_statistic, p_value = mannwhitneyu(small_contributors, large_contributors)

print(f"U-statistic: {u_statistic}, p-value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis. There is a significant difference in mean contributor lifespans between the groups.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in mean contributor lifespans between the groups.")
```

```python
import matplotlib.pyplot as plt

# Box plot comparison
plt.boxplot([small_contributors, large_contributors], labels=['Small Contributors', 'Large Contributors'])
plt.ylabel('Mean Lifespan')
plt.title('Comparison of Mean Lifespans')
plt.show()

# Histogram comparison
plt.hist([small_contributors, large_contributors], alpha=0.5, label=['Small Contributors', 'Large Contributors'])
plt.xlabel('Mean Lifespan')
plt.ylabel('Frequency')
plt.title('Distribution of Mean Lifespans')
plt.legend()
plt.show()
```
