```python
import csv


# Labels and keywords for filtering security issues
security_labels = ["security", "vulnerability", "vuln", "Bug", "help wanted", "conflicts", "linter error", "question"]
security_keywords = ["security", "vulnerability", "vuln", "cve", "exploit", "injection", "breach", "hack", "error", "broken", "question", "fix"]

# This will hold all the security issue data across repos
security_issue_data = []

for repo_name in repos:
    try:
        repo = g.get_repo(repo_name)
        print(f"Processing {repo_name} ...")
        repo_issues = {}

        # First, try to get issues using the specified security labels
        for label in security_labels:
            try:
                issues = repo.get_issues(state="all", labels=[label])
                for issue in issues:
                    if issue.number not in repo_issues:
                        created = issue.created_at
                        closed = issue.closed_at
                        resolution_time = (closed - created).total_seconds() / 3600 if closed else None
                        repo_issues[issue.number] = {
                            "repo": repo_name,
                            "issue_number": issue.number,
                            "state": issue.state,
                            "created_at": created,
                            "closed_at": closed,
                            "resolution_time_hours": resolution_time,
                            "title": issue.title
                        }
            except Exception as e:
                # Skip any label-specific issues errors
                continue

        if len(repo_issues) < 50:
            print(f"Less than 50 issues found by label for {repo_name}. Falling back to keyword search.")
            all_issues = repo.get_issues(state="all")
            for issue in all_issues:
                # Stop the fallback search once we have found 50 issues
                if len(repo_issues) >= 80:
                    break
                title_lower = issue.title.lower()
                if any(keyword in title_lower for keyword in security_keywords):
                    if issue.number not in repo_issues:
                        created = issue.created_at
                        closed = issue.closed_at
                        resolution_time = (closed - created).total_seconds() / 3600 if closed else None
                        repo_issues[issue.number] = {
                            "repo": repo_name,
                            "issue_number": issue.number,
                            "state": issue.state,
                            "created_at": created,
                            "closed_at": closed,
                            "resolution_time_hours": resolution_time,
                            "title": issue.title
                        }

        # Append all found issues for this repository to our master list
        for issue_data in repo_issues.values():
            security_issue_data.append(issue_data)

    except Exception as e:
        print(f"Error processing {repo_name}: {e}")
        continue

# Write all the collected security issues to a single CSV file
csv_filename = "security_issues.csv"
with open(csv_filename, "w", newline="", encoding="utf-8") as csvfile:
    fieldnames = ["repo", "issue_number", "state", "created_at", "closed_at", "resolution_time_hours", "title"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for issue in security_issue_data:
        writer.writerow(issue)

print(f"Security issues written to {csv_filename}")
```

```python
import pandas as pd
import numpy as np
from datetime import datetime

# Load the CSV file from the specified path
df = pd.read_csv('/content/security_issues.csv')

# Standardize column names: trim, convert to lowercase, and replace spaces with underscores
df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]

# Check if the 'repo' column exists
if 'repo' not in df.columns:
    raise KeyError("Column 'repo' not found in CSV. Available columns: " + str(df.columns))

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Trim whitespace from all text columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()

# Convert 'created_at' and 'closed_at' columns to datetime objects and remove timezone info
if 'created_at' in df.columns:
    df['created_at'] = pd.to_datetime(df['created_at'], errors='coerce')
    # Remove timezone info to make tz-naive
    df['created_at'] = df['created_at'].apply(lambda x: x.replace(tzinfo=None) if pd.notnull(x) else x)

if 'closed_at' in df.columns:
    # Replace empty strings with NaN, then convert to datetime and remove timezone info
    df['closed_at'] = df['closed_at'].replace('', np.nan)
    df['closed_at'] = pd.to_datetime(df['closed_at'], errors='coerce')
    df['closed_at'] = df['closed_at'].apply(lambda x: x.replace(tzinfo=None) if pd.notnull(x) else x)

# Set the current time to the specified value: "2025-03-15 10:45:00+00:00" and remove timezone info
current_time = pd.Timestamp("2025-03-15 10:45:00+00:00").tz_localize(None)

# For issues that are still open (i.e. where 'closed_at' is NaN), calculate open age in days.
df['open_age'] = np.where(
    df['closed_at'].isna(),
    (current_time - df['created_at']).dt.total_seconds() / (3600 * 24),  # convert seconds to days
    np.nan
)

# Group data by 'repo' to compute metrics per repository.
# - avg_open_age: average open age in days for issues that are still open.
# - avg_resolution_time: average resolution time from 'resolution_time_hours' converted from hours to days.
grouped = df.groupby('repo').agg(
    avg_open_age = ('open_age', lambda x: np.nanmean(x)),
    avg_resolution_time = ('resolution_time_hours', lambda x: np.nanmean(x) / 24)  # convert hours to days
)

# Count the total number of security issues per repository
issue_counts = df.groupby('repo').size().rename('num_security_issues')

# Merge the grouped metrics with the issue counts
repo_stats = grouped.merge(issue_counts, left_index=True, right_index=True)

# Calculate defect density for each repository using the repo_sizes dictionary.
# Estimated KLOC = (repo_size * 50) / 1000.0, and defect_density = num_security_issues / estimated_kloc if estimated_kloc > 0.
def compute_defect_density(repo, num_issues):
    size = repo_sizes.get(repo)  # repo_sizes dictionary must be defined externally
    if size is None:
        return None
    estimated_kloc = (size * 50) / 1000.0
    if estimated_kloc > 0:
        return num_issues / estimated_kloc
    else:
        return None

repo_stats['defect_density'] = repo_stats.apply(
    lambda row: compute_defect_density(row.name, row['num_security_issues']),
    axis=1
)

# Reset index to convert the 'repo' index into a column and rename it to 'repo_name'
repo_stats = repo_stats.reset_index().rename(columns={'repo': 'repo_name'})

# Select the desired columns: repo_name, avg_open_age, avg_resolution_time, and defect_density
final_df = repo_stats[['repo_name', 'avg_open_age', 'avg_resolution_time', 'defect_density']]

final_df['avg_open_age'] = final_df['avg_open_age'].fillna(0)


# Save the results to a CSV file
final_df.to_csv('repo_security_metrics.csv', index=False)

print(final_df)
```

```python
import pandas as pd

# Load both CSV files
avg_df = pd.read_csv('/content/avg_lifespan_per_repo.csv')
sec_df = pd.read_csv('/content/repo_security_metrics.csv')

# Standardize repository names: lower-case and strip extra spaces
avg_df['repo_name'] = avg_df['repo_name'].str.lower().str.strip()
sec_df['repo_name'] = sec_df['repo_name'].str.lower().str.strip()

# Create a mapping for avg_df repo names:
# If the name already contains a '/', we assume it's in author/repo format.
# Otherwise, try to find a repo in sec_df that ends with '/' + repo_name.
mapping = {}
for repo in avg_df['repo_name']:
    if '/' in repo:
        mapping[repo] = repo
    else:
        # Find matches in sec_df whose repo name ends with '/' + repo
        matches = sec_df[sec_df['repo_name'].str.endswith('/' + repo)]
        if not matches.empty:
            # If multiple matches, take the first one.
            mapping[repo] = matches.iloc[0]['repo_name']
        else:
            mapping[repo] = repo

# Update the avg_df 'repo_name' using the mapping
avg_df['repo_name'] = avg_df['repo_name'].map(mapping)

# Find common repositories between the two CSVs
common_repos = set(avg_df['repo_name']).intersection(set(sec_df['repo_name']))

print("Number of similar repositories:", len(common_repos))
print("Common repositories:", common_repos)

# Optionally, save the updated avg_df to a new CSV file with standardized repo names.
avg_df.to_csv('avg_lifespan_per_repo_standardized.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy.stats import pearsonr, spearmanr
from mpl_toolkits.mplot3d import Axes3D  # For 3D plotting

# -------------------------------------------------------------------
# 1. Load and Preprocess Data
# -------------------------------------------------------------------

# ---- Contributor Data ----
# Load contributor data from the provided CSV file
df_contributors = pd.read_csv("/content/github_fetched_contributor_data.xlsx - Sheet1.csv")
print("Contributor Data Columns:")
print(df_contributors.columns)

# Set the date column names based on the printed columns
first_date_col = "First Contribution"
last_date_col = "Last Contribution"

# Convert the date columns to datetime
df_contributors[first_date_col] = pd.to_datetime(df_contributors[first_date_col], errors='coerce')
df_contributors[last_date_col] = pd.to_datetime(df_contributors[last_date_col], errors='coerce')

# Remove timezone information
df_contributors[first_date_col] = df_contributors[first_date_col].dt.tz_localize(None)
df_contributors[last_date_col] = df_contributors[last_date_col].dt.tz_localize(None)

# We'll use the provided "Lifespan (days)" as the developer tenure measure.
# Rename the repository column to "repo" for consistency.
df_contributors.rename(columns={"Repository Name": "repo"}, inplace=True)
df_contributors["repo"] = df_contributors["repo"].str.lower().str.strip()

# ---- Security Metrics Data ----
df_security_metrics = pd.read_csv("/content/repo_security_metrics.csv")
# If repository column is named "repo_name", rename it to "repo"
if "repo_name" in df_security_metrics.columns:
    df_security_metrics.rename(columns={"repo_name": "repo"}, inplace=True)
df_security_metrics["repo"] = df_security_metrics["repo"].str.lower().str.strip()

# ---- Security Issues Data ----
df_security_issues = pd.read_csv("/content/security_issues.csv", parse_dates=["created_at", "closed_at"])
df_security_issues["created_at"] = df_security_issues["created_at"].dt.tz_localize(None)
df_security_issues["closed_at"] = df_security_issues["closed_at"].replace('', np.nan)
df_security_issues["closed_at"] = df_security_issues["closed_at"].apply(
    lambda x: pd.to_datetime(x).tz_localize(None) if pd.notnull(x) else x
)

# -------------------------------------------------------------------
# 2. Aggregate Developer Tenure and Compute Commit Frequency
# -------------------------------------------------------------------
# Aggregate developer tenure metrics per repository from contributor data.
# Use the provided "Lifespan (days)" as the tenure measure.
agg_tenure = df_contributors.groupby("repo").agg(
    avg_tenure_days=("Lifespan (days)", "mean"),
    median_tenure_days=("Lifespan (days)", "median"),
    num_contributors=("Contributor", "nunique"),
    total_contributions=("Total Contributions", "sum")  # Using Total Contributions from the data
).reset_index()

print("Aggregated Developer Tenure Metrics:")
print(agg_tenure.head())

# Merge security metrics with contributor tenure info on "repo"
df_merged = pd.merge(df_security_metrics, agg_tenure, on="repo", how="left")
print("\nMerged Data (Security Metrics + Developer Tenure):")
print(df_merged.head())

# Calculate open issue age (in days) for each security issue
current_date = pd.Timestamp.now()
df_security_issues["open_issue_age_days"] = (current_date - df_security_issues["created_at"]).dt.days

# Calculate commit frequency per repository from contributor data.
# We use the "First Contribution" and "Last Contribution" to define activity period,
# and "Total Contributions" as the total number of commits.
commit_freq = df_contributors.groupby("repo").agg(
    earliest_commit=(first_date_col, "min"),
    latest_commit=(last_date_col, "max"),
    total_contributions=("Total Contributions", "sum")
).reset_index()

commit_freq["earliest_commit"] = commit_freq["earliest_commit"].dt.tz_localize(None)
commit_freq["latest_commit"] = commit_freq["latest_commit"].dt.tz_localize(None)
commit_freq["days_active"] = (commit_freq["latest_commit"] - commit_freq["earliest_commit"]).dt.days.replace(0, 1) + 1
commit_freq["commits_per_day"] = commit_freq["total_contributions"] / commit_freq["days_active"]

print("\nOverall Commit Frequency per Repository:")
print(commit_freq.head())

# Merge commit frequency into the main merged data
df_merged_extended = pd.merge(df_merged, commit_freq[["repo", "commits_per_day"]], on="repo", how="left")
print("\nMerged Data with Commit Frequency:")
print(df_merged_extended.head())

# -------------------------------------------------------------------
# 3. Extended Analysis
# -------------------------------------------------------------------
# --- Multiple Regression Analysis ---
# Predict defect_density using multiple predictors:
# We'll use: avg_tenure_days, commits_per_day, avg_resolution_time_hours, and num_contributors.
features = ["avg_tenure_days", "commits_per_day", "avg_resolution_time_hours", "num_contributors"]
available_features = [f for f in features if f in df_merged_extended.columns]

reg_data = df_merged_extended.dropna(subset=["defect_density"] + available_features).copy()
X = reg_data[available_features]
y = reg_data["defect_density"]

# Add constant term for intercept
X = sm.add_constant(X)

model_multi = sm.OLS(y, X).fit()
print("\n--- Multiple Regression: Defect Density ~ Multiple Predictors ---")
print(model_multi.summary())

# --- 3D Scatter Plot ---
# Visualize the interaction among avg_tenure_days, commits_per_day, and defect_density.
df_3d = reg_data.dropna(subset=["avg_tenure_days", "commits_per_day", "defect_density"]).copy()

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(
    df_3d["avg_tenure_days"],
    df_3d["commits_per_day"],
    df_3d["defect_density"],
    c='b', alpha=0.6, s=50
)
ax.set_xlabel("Avg Developer Tenure (days)")
ax.set_ylabel("Commits per Day")
ax.set_zlabel("Defect Density")
ax.set_title("3D Scatter: Tenure, Commits/Day, & Defect Density")
plt.show()

# --- Additional Visualizations ---
# Scatter plot: Developer Tenure vs. Defect Density
plt.figure(figsize=(8,6))
sns.scatterplot(data=df_merged_extended, x="avg_tenure_days", y="defect_density", s=100)
plt.xlabel("Average Developer Tenure (days)")
plt.ylabel("Defect Density (issues per estimated KLOC)")
plt.title("Developer Tenure vs. Defect Density")
plt.show()

# Scatter plot: Developer Tenure vs. Average Resolution Time (if available)
if "avg_resolution_time_hours" in df_merged_extended.columns:
    plt.figure(figsize=(8,6))
    sns.scatterplot(data=df_merged_extended, x="avg_tenure_days", y="avg_resolution_time_hours", s=100)
    plt.xlabel("Average Developer Tenure (days)")
    plt.ylabel("Average Resolution Time (hours)")
    plt.title("Developer Tenure vs. Average Resolution Time")
    plt.show()

# Boxplot: Developer Tenure vs. Number of Contributors
if "num_contributors" in df_merged_extended.columns:
    plt.figure(figsize=(8,6))
    sns.boxplot(data=df_merged_extended, x="num_contributors", y="avg_tenure_days")
    plt.xlabel("Number of Contributors")
    plt.ylabel("Average Developer Tenure (days)")
    plt.title("Developer Tenure vs. Number of Contributors")
    plt.show()

# Pairplot for selected variables
vars_for_pairplot = ["avg_tenure_days", "defect_density", "commits_per_day"]
if "avg_resolution_time_hours" in df_merged_extended.columns:
    vars_for_pairplot.append("avg_resolution_time_hours")
sns.pairplot(df_merged_extended[vars_for_pairplot].dropna(), corner=True)
plt.suptitle("Pair Plot of Developer and Security Metrics", y=1.02)
plt.show()

# --- Simple Linear Regressions for Comparison ---
# Regression: Defect Density ~ Avg Developer Tenure
reg_data_simple = df_merged_extended.dropna(subset=["avg_tenure_days", "defect_density"]).copy()
X_simple = sm.add_constant(reg_data_simple["avg_tenure_days"])
model_simple = sm.OLS(reg_data_simple["defect_density"], X_simple).fit()
print("\nRegression Analysis: Defect Density ~ Avg Developer Tenure")
print(model_simple.summary())

# Regression: Average Resolution Time ~ Avg Developer Tenure (if available)
if "avg_resolution_time_hours" in df_merged_extended.columns:
    reg_data_res = df_merged_extended.dropna(subset=["avg_tenure_days", "avg_resolution_time_hours"]).copy()
    X_res = sm.add_constant(reg_data_res["avg_tenure_days"])
    model_res = sm.OLS(reg_data_res["avg_resolution_time_hours"], X_res).fit()
    print("\nRegression Analysis: Average Resolution Time ~ Avg Developer Tenure")
    print(model_res.summary())
```

```python
import pandas as pd
from scipy.stats import spearmanr, mannwhitneyu

# Load datasets
repo_metrics_df = pd.read_csv("repo_security_metrics (1).csv")
avg_lifespan_df = pd.read_csv("avg_lifespan_per_repo (1).csv")

# Merge datasets on 'repo_name' to align them for analysis
merged_df = pd.merge(avg_lifespan_df, repo_metrics_df, on="repo_name", how="inner")

# Selecting numerical columns for analysis
numeric_columns = ['avg_lifespan', 'avg_open_age', 'avg_resolution_time', 'defect_density']

# Spearman correlation analysis
spearman_results = merged_df[numeric_columns].corr(method="spearman")

# Mann-Whitney U Test (Using the median of 'avg_lifespan' for grouping)
median_split = merged_df["avg_lifespan"].median()
merged_df["lifespan_category"] = merged_df["avg_lifespan"] > median_split  # True/False groups

# Perform Mann-Whitney U test for each numerical column
mann_whitney_results = {}
for col in ['avg_open_age', 'avg_resolution_time', 'defect_density']:
    group1 = merged_df[merged_df["lifespan_category"] == True][col]
    group2 = merged_df[merged_df["lifespan_category"] == False][col]

    stat, p_value = mannwhitneyu(group1, group2, alternative='two-sided')
    mann_whitney_results[col] = {"U-statistic": stat, "p-value": p_value}

mann_whitney_df = pd.DataFrame.from_dict(mann_whitney_results, orient="index")

# Display results in local execution
print("\nSpearman Correlation Results:\n", spearman_results)
print("\nMann-Whitney U Test Results:\n", mann_whitney_df)

```
